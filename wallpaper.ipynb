```python
canvas = Image.new(color="#F8BBD0", mode="RGBA", size=(1200,480))
k = Image.open("/home/vinay/sneha/wallpaper/ic_whatshot_black_48dp.png")
k.mode = "RGBA"
k = k.resize(size=(18,18), resample=Image.ANTIALIAS)
for i in range(int(canvas.size[0]/k.size[0])):
    x1 = i*k.size[0]
    x2 = k.size[0]+x1
    for j in range(int(canvas.size[1]/k.size[1])):
        y1 = j*k.size[1]
        y2 = y1+k.size[1]
        try:
            canvas.paste(box=(x1, y1, x2, y2), im = k, mask= k)
        except:
            print(i)
            raise

canvas
```
